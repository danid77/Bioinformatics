## 정석

1. 컬럼을 데이터 0번 index에 추가한다
2. 컬럼명을 바꾼다. 
3. Phenotype_value을 '1' = control, '2' = case로 치환한다. 
4. 0번 index로 컬럼명을 바꾸고 0번 index는 제거한다. 
5. 인덱스 없이 출력한다. 

```python
# 1.컬럼을 데이터 0번 index에 추가한다
col = pd.DataFrame([[df.columns[0], int(df.columns[1])]], columns = df.columns)
df_col = pd.concat([col, df])
df_col = df_col.reset_index()
df_col = df_col.drop(columns='index', axis=1)

# 2. 컬럼명을 바꾼다
df_col.columns = ['FAMID', 'Phenotype_value']
df_col
```

```python
df_col[df_col['Phenotype_value'] == 2]
```

```python
# 3. Phenotype_value을 '1' = control, '2' = case로 치환한다.
df_col.loc[df_col[df_col.loc[:]['Phenotype_value'] == 2].index, 'Phenotype_value'] = 'Case'
df_col.loc[df_col[df_col.loc[:]['Phenotype_value'] == 1].index, 'Phenotype_value'] = 'control'
```

```python
# 숫자 차환 확인
df_col['Phenotype_value'].value_counts()
```

```python
# 4. 0번 index로 컬럼명을 바꾸고 0번 index는 제거한다. 
df_col = df_col.rename(columns=df_col.loc[0])
df_col = df_col[1:]
df_col
```

```python
# 5. 인덱스 없이 출력한다. 
df_col.to_csv('./data_sample_fnl_2.txt', sep='\t', index=False)
```
