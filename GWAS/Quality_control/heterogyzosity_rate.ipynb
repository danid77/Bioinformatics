```python
import pandas as pd
import numpy as np
```

```python
# 파일 불러오기
het = pd.read_csv('korea_het.het', sep='\t', engine='python', encoding='cp949')
het
```

```python
het = het[['FID', 'IID', 'O(HOM)', 'N(NM)']]
```

```python
het
```

```python
het.columns = ['FID', 'IID', 'O', 'N']
```

```python
het['Heterozygosity_rate'] = (het['N'] - het['O']) / het['N'] 
```

```python
het
```

```python
het[['FID', 'IID', 'Heterozygosity_rate']]
```

```python
print(round((het['N'][0] - het['O'][0]) / het['N'][0], 6))
print(round((het['N'][1] - het['O'][1]) / het['N'][1], 6)) 
print(round((het['N'][2118] - het['O'][2118]) / het['N'][2118], 6))
```

```python
het['Heterozygosity_rate'].std(axis=0) * 3
```

```python
het['Heterozygosity_rate'].mean() - (het['Heterozygosity_rate'].std(axis=0) * 3)
```

```python
het['Heterozygosity_rate'].mean() + (het['Heterozygosity_rate'].std(axis=0) * 3)
```

```python
het_del = het[(het['Heterozygosity_rate'] >= het['Heterozygosity_rate'].mean() + (het['Heterozygosity_rate'].std(axis=0) * 3)) | (het['Heterozygosity_rate'] <= het['Heterozygosity_rate'].mean() - (het['Heterozygosity_rate'].std(axis=0) * 3))]
```
