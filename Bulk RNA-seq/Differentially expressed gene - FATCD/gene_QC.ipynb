## DEG에서 미토콘드리아 gene이 있는지 확인

```python
CD_Con = pd.read_csv('CD_Con.txt', sep=',', engine='python', encoding='cp949')
CrF_Con = pd.read_csv('CrF_Con.txt', sep=',', engine='python', encoding='cp949')
CrF_CD = pd.read_csv('CrF_CD.txt', sep=',', engine='python', encoding='cp949')

CD_Con = CD_Con.rename(columns={'Unnamed: 0' : 'gene'})
CrF_Con = CrF_Con.rename(columns={'Unnamed: 0' : 'gene'})
CrF_CD = CrF_CD.rename(columns={'Unnamed: 0' : 'gene'})

CD_Con_deg = CD_Con[(CD_Con['padj'] < 0.05) & (abs(CD_Con['log2FoldChange']) > 1.5)]
CrF_Con_deg = CrF_Con[(CrF_Con['padj'] < 0.05) & (abs(CrF_Con['log2FoldChange']) > 1.5)]
CrF_CD_deg = CrF_CD[(CrF_CD['padj'] < 0.05) & (abs(CrF_CD['log2FoldChange']) > 1.5)]

display(CD_Con_deg)
display(CrF_Con_deg)
display(CrF_CD_deg)
```

```python
display(gtf_type[gtf_type['gene'].isin(list(CD_Con_deg['gene']))])
display(gtf_type[gtf_type['gene'].isin(list(CrF_Con_deg['gene']))])
display(gtf_type[gtf_type['gene'].isin(list(CrF_CD_deg['gene']))])
```
